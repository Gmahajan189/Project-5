```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Check the column names
print(data.columns)

# Aggregating session frequency, total duration, and total traffic per user (MSISDN/Number)
engagement_metrics = data.groupby('MSISDN/Number').agg({
    'Bearer Id': 'count',       # Session Frequency
    'Dur. (ms)': 'sum',         # Total Duration
    'Total DL (Bytes)': 'sum',  # Total Download Data
    'Total UL (Bytes)': 'sum'   # Total Upload Data
}).reset_index()

# Rename columns for clarity
engagement_metrics.columns = ['MSISDN/Number', 'Sessions Frequency', 'Total Duration', 'Total DL', 'Total UL']

# Calculate total traffic (DL + UL)
engagement_metrics['Total Traffic'] = engagement_metrics['Total DL'] + engagement_metrics['Total UL']

# Display top 10 customers per engagement metric
top_sessions = engagement_metrics.sort_values('Sessions Frequency', ascending=False).head(10)
top_duration = engagement_metrics.sort_values('Total Duration', ascending=False).head(10)
top_traffic = engagement_metrics.sort_values('Total Traffic', ascending=False).head(10)

print("Top 10 Customers by Sessions Frequency:")
print(top_sessions)

print("Top 10 Customers by Total Duration:")
print(top_duration)

print("Top 10 Customers by Total Traffic:")
print(top_traffic)

# Normalizing the metrics for k-means clustering
scaler = MinMaxScaler()
engagement_metrics[['Sessions Frequency', 'Total Duration', 'Total Traffic']] = scaler.fit_transform(
    engagement_metrics[['Sessions Frequency', 'Total Duration', 'Total Traffic']]
)

# Running k-means clustering with k=3
kmeans = KMeans(n_clusters=3, random_state=42)
engagement_metrics['Cluster'] = kmeans.fit_predict(engagement_metrics[['Sessions Frequency', 'Total Duration', 'Total Traffic']])

# Display cluster metrics
cluster_summary = engagement_metrics.groupby('Cluster').agg({
    'Sessions Frequency': ['min', 'max', 'mean', 'sum'],
    'Total Duration': ['min', 'max', 'mean', 'sum'],
    'Total Traffic': ['min', 'max', 'mean', 'sum']
}).reset_index()

print(cluster_summary)

# Visualizing the clusters
sns.scatterplot(data=engagement_metrics, x='Sessions Frequency', y='Total Traffic', hue='Cluster', palette='viridis')
plt.title('Clusters of User Engagement')
plt.show()

# Elbow method to find optimal k value
inertia = []
k_values = range(1, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(engagement_metrics[['Sessions Frequency', 'Total Duration', 'Total Traffic']])
    inertia.append(kmeans.inertia_)

plt.plot(k_values, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method to Determine Optimal k')
plt.show()


```

```python
# Aggregating session frequency, total duration, and total traffic per user (MSISDN)
engagement_metrics = data.groupby('MSISDN/Number').agg({
    'Bearer Id': 'count',       # Session Frequency
    'Dur. (ms)': 'sum',         # Total Duration
    'Total DL (Bytes)': 'sum',  # Total Download Data
    'Total UL (Bytes)': 'sum'   # Total Upload Data
}).reset_index()

# Renaming columns for clarity
engagement_metrics.columns = ['MSISDN', 'Sessions Frequency', 'Total Duration', 'Total DL', 'Total UL']

# Calculating total traffic (DL + UL)
engagement_metrics['Total Traffic'] = engagement_metrics['Total DL'] + engagement_metrics['Total UL']

# Display top 10 customers per engagement metric
top_sessions = engagement_metrics.sort_values('Sessions Frequency', ascending=False).head(10)
top_duration = engagement_metrics.sort_values('Total Duration', ascending=False).head(10)
top_traffic = engagement_metrics.sort_values('Total Traffic', ascending=False).head(10)

print("Top 10 Customers by Sessions Frequency:")
print(top_sessions)

print("Top 10 Customers by Total Duration:")
print(top_duration)

print("Top 10 Customers by Total Traffic:")
print(top_traffic)
```

```python
# Selecting features for normalization
features = ['Sessions Frequency', 'Total Duration', 'Total Traffic']

# Normalize the features
scaler = StandardScaler()
engagement_metrics_normalized = scaler.fit_transform(engagement_metrics[features])

# K-means clustering with k=3
kmeans = KMeans(n_clusters=3, random_state=42)
engagement_metrics['Cluster'] = kmeans.fit_predict(engagement_metrics_normalized)

# Show cluster assignments
print(engagement_metrics[['MSISDN', 'Cluster']].head())
```

```python
# Compute min, max, average, and total for each cluster
cluster_stats = engagement_metrics.groupby('Cluster').agg({
    'Sessions Frequency': ['min', 'max', 'mean', 'sum'],
    'Total Duration': ['min', 'max', 'mean', 'sum'],
    'Total Traffic': ['min', 'max', 'mean', 'sum']
})

print(cluster_stats)

# Visualize cluster statistics
cluster_stats.plot(kind='bar', subplots=True, layout=(4, 3), figsize=(12, 10), sharex=False)
plt.suptitle('Cluster Statistics for Engagement Metrics')
plt.tight_layout()
plt.show()

```

**Interpretation:**

**Cluster 0**: Users in this cluster have low session frequencies but high variability in total duration and traffic. This cluster may represent casual users with varying engagement levels.

**Cluster 1:** Users in this cluster exhibit very high and consistent engagement metrics across all dimensions, suggesting a highly engaged user base with uniform usage patterns.

**Cluster 2**: Users here have moderate session frequency and significant variability in total duration and traffic. This cluster could represent users with a more varied engagement pattern compared to Cluster 1 but more consistent than Cluster 0.
